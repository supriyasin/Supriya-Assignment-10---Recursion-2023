```python
#<aside>
💡 **Question 1**

Given an integer `n`, return *`true` if it is a power of three. Otherwise, return `false`*.

An integer `n` is a power of three, if there exists an integer `x` such that `n == 3x`.

**Example 1:**

Input: n = 27
Output: true
Explanation: 27 = 33          
    
Example 2:     
Input: n = 0
Output: false
Explanation: There is no x where 3x = 0.
    
Example 3:        
Input: n = -1
Output: false
Explanation: There is no x where 3x = (-1).
 
<aside>

"""To determine if an integer is a power of three, we can use the following approach:

   1. Check if the given integer n is less than or equal to 0. If so, return false since powers of three must be positive 
      integers.
   2. While n is divisible by 3, divide n by 3.
   
   3. After the loop, if n is equal to 1, it means the original number was a power of three. Return true.
   
   4. Otherwise, return false because the original number was not a power of three."""

#Here's the implementation of the above algorithm in Python:

def isPowerOfThree(n):
    if n <= 0:
        return False
    
    while n % 3 == 0:
        n //= 3
    
    return n == 1
print(isPowerOfThree(27))  # Output: True
print(isPowerOfThree(0))   # Output: False
print(isPowerOfThree(-1))  # Output: False

#The function should return True for the input 27, False for the input 0, and False for the input -1, as stated in the examples.
```

```python
#Question 2
You have a list `arr` of all integers in the range `[1, n]` sorted in a strictly increasing order. Apply the following 
algorithm on `arr`:

- Starting from left to right, remove the first number and every other number afterward until you reach the end of the list.
- Repeat the previous step again, but this time from right to left, remove the rightmost number and every other number from 
  the remaining numbers.
- Keep repeating the steps again, alternating left to right and right to left, until a single number remains.

Given the integer `n`, return *the last number that remains in* `arr`.

**Example 1:**    Input: n = 9
Output: 6
Explanation:   
arr = [1, 2,3, 4,5, 6,7, 8,9]
arr = [2,4, 6,8]
arr = [2, 6]
arr = [6]

Example 2:     
Input: n = 1
Output: 1
    
"""To solve this problem efficiently, we can observe a pattern. Let's go through the steps of the algorithm for a few 
   different values of n:

   1. For n = 9:
      • Initially: [1, 2, 3, 4, 5, 6, 7, 8, 9]
      • After removing every other number from left to right: [2, 4, 6, 8]
      • After removing every other number from right to left: [2, 6]
      • After removing every other number from left to right: [6]
      • The last number remaining is 6.
      
   2. For n = 8:
      • Initially: [1, 2, 3, 4, 5, 6, 7, 8]
      • After removing every other number from left to right: [2, 4, 6, 8]
      • After removing every other number from right to left: [2, 6]
      • The last number remaining is 6.
      
   3. For n = 7:
      • Initially: [1, 2, 3, 4, 5, 6, 7]
      • After removing every other number from left to right: [2, 4, 6]
      • After removing every other number from right to left: [2]
      • The last number remaining is 2.
      
 Observations:
 • When n is even, the last number remaining after one iteration is the same as when n/2 is used as the input.
 • When n is odd, the last number remaining after one iteration is the same as when n-1 is used as the input.
 • We can recursively apply this logic until we reach the base case of n = 1."""

#Based on these observations, we can implement a recursive solution to find the last number remaining:

def lastRemaining(n: int) -> int:
    if n == 1:
        return 1
    elif n % 2 == 0:
        return lastRemaining(n // 2)
    else:
        return lastRemaining(n - 1)
print(lastRemaining(9))  # Output: 6
print(lastRemaining(1))  # Output: 1

#The output matches the expected results.
```

```python
#<aside>
💡 **Question 4**

Given a string calculate length of the string using recursion.

**Examples:**

Input : str = "abcd"
Output :4

Input : str = "GEEKSFORGEEKS"
Output :13

<aside>

"""To calculate the length of a string using recursion, we can follow a simple approach. Let's define a recursive function 
   called calculateLength that takes the string as input."""

#Here's the implementation in Python:

def calculateLength(string):
    if string == "":
        return 0
    else:
        return 1 + calculateLength(string[1:])

# Test the function
string1 = "abcd"
print(calculateLength(string1))  # Output: 4

string2 = "GEEKSFORGEEKS"
print(calculateLength(string2))  # Output: 13

"""In this code, the calculateLength function takes the string as input. We have two cases:

   1. Base case: If the string is empty, i.e., string == "", we return 0 because there are no characters in the string.
   
   2. Recursive case: If the string is not empty, we remove the first character from the string using slicing (string[1:])
      and make a recursive call to calculateLength with the updated string. We add 1 to the result of the recursive call 
      because we are counting the current character as well.
      
   The recursive calls continue until the string becomes empty, and the function starts returning the accumulated length as
   the recursive calls unwind.

   By following this approach, we can calculate the length of a string using recursion."""
```

```python
#<aside>
💡 **Question 5**

We are given a string S, we need to find count of all contiguous substrings starting and ending with same character.

**Examples :**

Input  : S = "abcab"
Output : 7
There are 15 substrings of "abcab"
a, ab, abc, abca, abcab, b, bc, bca
bcab, c, ca, cab, a, ab, b
Out of the above substrings, there
are 7 substrings : a, abca, b, bcab,
c, a and b.

Input  : S = "aba"
Output : 4
The substrings are a, b, a and aba

</aside>

"""To find the count of all contiguous substrings in a string S that start and end with the same character, we can use the 
   following approach:

   1. Initialize a variable count to 0 to keep track of the count of substrings.
   2. Iterate over each character in the string S.
   3. For each character at index i, consider it as the starting character of a substring. Then, iterate from i to the end
      of the string, considering each character as the ending character of the substring.
   4. If the starting and ending characters are the same, increment the count variable.
   5. After iterating over all possible substrings, return the final value of count."""

#Here's the implementation of this approach in Python:

def countContiguousSubstrings(S):
    count = 0
    for i in range(len(S)):
        for j in range(i, len(S)):
            if S[i] == S[j]:
                count += 1
    return count

# Test the function
S1 = "abcab"
print(countContiguousSubstrings(S1))  # Output: 7

S2 = "aba"
print(countContiguousSubstrings(S2))  # Output: 4

"""In this code, we iterate over each character in the string using two nested loops. The outer loop represents the starting 
   index of the substring, and the inner loop represents the ending index of the substring. We compare the characters at these
   indices, and if they are the same, we increment the count variable.

   By considering all possible starting and ending indices, we count all the contiguous substrings that start and end with the
   same character.

   The time complexity of this approach is O(n^2), where n is the length of the string S."""
```

```python
#<aside>
💡 **Question 6**

The [tower of Hanoi](https://en.wikipedia.org/wiki/Tower_of_Hanoi) is a famous puzzle where we have three rods and **N** disks. 
The objective of the puzzle is to move the entire stack to another rod. You are given the number of discs **N**. Initially,
these discs are in the rod 1. You need to print all the steps of discs movement so that all the discs reach the 3rd rod. Also, 
you need to find the total moves.**Note:** The discs are arranged such that the **top disc is numbered 1** and the 
    **bottom-most disc is numbered N**. Also, all the discs have **different sizes** and a bigger disc **cannot** be put on 
    the top of a smaller disc. Refer the provided link to get a better clarity about the puzzle.

**Example 1:**
Input:
N = 2
Output:
move disk 1 from rod 1 to rod 2
move disk 2 from rod 1 to rod 3
move disk 1 from rod 2 to rod 3
3
Explanation:For N=2 , steps will be
as follows in the example and total
3 steps will be taken.

</aside>  

"""To solve the Tower of Hanoi puzzle and print all the steps for moving the discs from one rod to another, we can use a 
   recursive approach."""

#Here's how we can implement it in Python:

def towerOfHanoi(n, source, destination, auxiliary):
    if n == 1:
        print("Move disk 1 from rod", source, "to rod", destination)
        return 1

    moves = 0

    moves += towerOfHanoi(n-1, source, auxiliary, destination)
    print("Move disk", n, "from rod", source, "to rod", destination)
    moves += 1
    moves += towerOfHanoi(n-1, auxiliary, destination, source)

    return moves

# Test the function
N = 2
total_moves = towerOfHanoi(N, 1, 3, 2)
print("Total moves:", total_moves)

"""In this code, the towerOfHanoi function takes four parameters:

   • n: The number of discs to be moved.
   • source: The rod number from which the discs should be moved.
   • destination: The rod number to which the discs should be moved.
   • auxiliary: The rod number that can be used as an auxiliary to facilitate the movement.
   
  The base case occurs when n is equal to 1. In this case, we directly move the only disk from the source rod to the 
  destination rod and return 1 as the number of moves.

  In the recursive part, we perform the following steps:
  
   1. Move n-1 discs from the source rod to the auxiliary rod using the destination rod as the auxiliary.
   2. Print the step for moving the nth disk from the source rod to the destination rod.
   3. Increment the moves counter.
   4. Move the n-1 discs from the auxiliary rod to the destination rod using the source rod as the auxiliary.
   
   We accumulate the number of moves by adding the number of moves from the recursive calls.

   By following this recursive approach, we can print all the steps for moving the discs and find the total number of moves
   required to solve the Tower of Hanoi puzzle."""
```

```python
#<aside>
💡 **Question 7**

Given a string **str**, the task is to print all the permutations of **str**. A **permutation** is an arrangement of all or
part of a set of objects, with regard to the order of the arrangement. For instance, the words ‘bat’ and ‘tab’ represents two 
distinct permutation (or arrangements) of a similar three letter word.

**Examples:**

> nput: str = “cd”
> 
> 
> **Output:** cd dc
> 
> **Input:** str = “abb”
> 
> **Output:** abb abb bab bba bab bba
>
<aside>

"""To print all permutations of a given string str, we can use a recursive approach."""

# Here's how we can implement it in Python:

def permute(str):
    # Convert the string to a list to make it mutable
    str_list = list(str)
    permute_helper(str_list, 0, len(str) - 1)

def permute_helper(str_list, left, right):
    if left == right:
        # Base case: reached the end of the string, print the permutation
        print("".join(str_list))
    else:
        for i in range(left, right + 1):
            # Swap the characters at indices left and i
            str_list[left], str_list[i] = str_list[i], str_list[left]

            # Recursively generate permutations for the remaining characters
            permute_helper(str_list, left + 1, right)

            # Restore the original order by swapping the characters back
            str_list[left], str_list[i] = str_list[i], str_list[left]

# Test the function
str1 = "cd"
permute(str1)

str2 = "abb"
permute(str2)

"""In this code, we define the permute function that takes the string str as input. We convert the string to a list (str_list) 
   to make it mutable and pass it to the permute_helper function.

   The permute_helper function performs the following steps:
   
   • If left is equal to right, it means we have reached the end of the string, and we print the permutation.
   • Otherwise, we iterate over all indices from left to right. For each index i, we swap the characters at indices left
     and i to generate a new permutation.
   • After swapping, we recursively generate permutations for the remaining characters by making a recursive call to 
     permute_helper with left + 1 as the new value of left.
   • Once we have generated all permutations for the remaining characters, we restore the original order by swapping the 
     characters back.
     
 By following this recursive approach and generating all possible swaps, we can print all the permutations of the given 
 string."""
```

```python
#<aside>
💡 **Question 8**

Given a string, count total number of consonants in it. A consonant is an English alphabet character that is not vowel 
(a, e, i, o and u). Examples of constants are b, c, d, f, and g.

**Examples :**

Input : abc de
Output : 3
There are three consonants b, c and d.

Input : geeksforgeeks portal
Output : 12
    
</aside> 

"""To count the total number of consonants in a given string, we can iterate over each character in the string and check if 
   it is a consonant."""

#Here's how we can implement it in Python:

def countConsonants(string):
    consonants = "bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ"
    count = 0
    for char in string:
        if char in consonants:
            count += 1
    return count

# Test the function
string1 = "abc de"
print(countConsonants(string1))  # Output: 3

string2 = "geeksforgeeks portal"
print(countConsonants(string2))  # Output: 12


"""In this code, the countConsonants function takes the string as input. We define a string consonants that contains all the 
   consonant characters in both lowercase and uppercase.

   We initialize a variable count to 0 to keep track of the number of consonants. Then, we iterate over each character in the 
   string and check if it is present in the consonants string. If it is, we increment the count variable.

   Finally, we return the value of count, which represents the total number of consonants in the given string.

   By using this approach, we can count the number of consonants in any given string."""
```
